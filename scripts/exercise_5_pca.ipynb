## Exercise 5

Test the PCA class in a jupyter notebook using the iris.csv dataset (classificação).

```python
from si.io.csv_file import read_csv
from si.decomposition.pca import PCA
```

```python
# Loading the dataset iris.csv
iris = read_csv('../datasets/iris/iris.csv', features=True, label=True)

print('Original X shape:', iris.X.shape)
print('Original features:', iris.features)
```

```python
pca = PCA(n_components=2)
```

```python
print("\nMean of the data:\n", pca.mean)
print("\nPrincipal Components (rows):\n", pca.components)
print("\nExplained Variance:\n", pca.explained_variance)
```

```python
pca.fit(iris)
iris_pca=pca.transform(iris)
```

```python
print('Transformed X shape:', iris_pca.X.shape)
print('PCA features', iris_pca.features)
print('Explained variance ratio:', pca.explained_variance)
print('Principal components matrix shape:', pca.components.shape)
```
